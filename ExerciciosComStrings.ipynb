**Tamanho de strings. Faça um programa que leia 2 strings e informe o conteúdo delas seguido do seu comprimento. Informe também se as duas strings possuem o mesmo comprimento e são iguais ou diferentes no conteúdo.**


```
Compara duas strings
String 1: Brasil Hexa 2006
String 2: Brasil! Hexa 2006!
Tamanho de "Brasil Hexa 2006": 16 caracteres
Tamanho de "Brasil! Hexa 2006!": 18 caracteres
As duas strings são de tamanhos diferentes.
As duas strings possuem conteúdo diferente.
```



````python
string1 = input('Digite a informação 1: ')
string2 = input('Digite a informação 2: ')

tamanho1 = len(string1)
tamanho2 = len(string2)

print(f'String 1: {string1} - Tamanho: {tamanho1}')
print(f'String 1: {string1} - Tamanho: {tamanho2}')

if tamanho1 == tamanho2:
  print('As duas strings possuem o mesmo tamanho')
else: 
  print('As duas strings possuem tamanhos diferentes')

if string1 == string2:
  print('As duas strings possuem o mesmo conteúdo')
else:
  print('As duas strings possuem conteúdos diferentes')
````

**Palíndromo. Um palíndromo é uma seqüência de caracteres cuja leitura é idêntica se feita da direita para esquerda ou vice−versa. Por exemplo: OSSO e OVO são palíndromos. Em textos mais complexos os espaços e pontuação são ignorados. A frase SUBI NO ONIBUS é o exemplo de uma frase palíndroma onde os espaços foram ignorados. Faça um programa que leia uma seqüência de caracteres, mostre−a e diga se é um palíndromo ou não.**

````python
frase = input('Digite uma frase: ').replace(' ', '')

frase2 = frase[::-1]

if frase == frase2:
  print(f'{frase} = {frase2}\nÉ palíndromo')
else:
  print(f'{frase} != {frase2}\nNão é palíndromo')
````

**Número por extenso. Escreva um programa que solicite ao usuário a digitação de um número até 99 e imprima-o na tela por extenso.**

````python
def numero_extenso(n):
  numeros = {0: 'zero', 1: 'um', 2: 'dois', 3: 'três', 4: 'quatro', 5: 'cinco',
          6: 'seis', 7: 'sete', 8: 'oito', 9: 'nove', 10: 'dez',
          11: 'onze', 12: 'doze', 13: 'treze', 14: 'quatorze', 15: 'quinze',
          16: 'dezesseis', 17: 'dezessete', 18: 'dezoito', 19: 'dezenove', 20: 'vinte',
          30: 'trinta', 40: 'quarenta', 50: 'cinquenta', 60: 'sessenta', 70: 'setenta',
          80: 'oitenta', 90: 'noventa'}

  if n > 99 or n < 0:
    print('Número inválido')

  if n < 20 or n % 10 == 0:
    return numeros.get(n)

  decimal = n // 10 * 10
  unidade = n % 10
  extenso = (f'{numeros.get(decimal)} e {numeros.get(unidade)}')
  return extenso

n = int(input('Digite um número: '))
print(numero_extenso(n))
````

**Jogo de Forca. Desenvolva um jogo da forca. O programa terá uma lista de palavras lidas de um arquivo texto e escolherá uma aleatoriamente. O jogador poderá errar 6 vezes antes de ser enforcado.**

```
Digite uma letra: A
-> Você errou pela 1ª vez. Tente de novo!

Digite uma letra: O
A palavra é: _ _ _ _ O

Digite uma letra: E
A palavra é: _ E _ _ O

Digite uma letra: S
-> Você errou pela 2ª vez. Tente de novo!
```



````python
import random

palavras = open('/content/palavras.txt', 'r')
palavras = palavras.readlines()

palavra = random.choice(palavras).lower()
palavra = palavra.strip('\n')
forca = ['_' for i in range (len(palavra))]
erros = 0
ganhou = False

def imprime_forca(forca):
  print(' '.join(forca).upper())
  print('')

while erros <= 6 and not ganhou:
  imprime_forca(forca)
  chute = str(input('Digite uma letra: ')).lower()

  if chute not in palavra:
    erros += 1

    if erros > 6:
      ganhou == False
      break

    print(f'Você errou pela {erros}ª vez. Tente de novo!')
    continue

  for i, letra in enumerate(palavra):
    if letra == chute:
      forca[i] = chute

  if '_' not in forca:
    ganhou = True

if ganhou:
  print('Você ganhou!')
else:
  print('O jogo acabou, você perdeu!')

imprime_forca(forca)
````

**Jogo da palavra embaralhada. Desenvolva um jogo em que o usuário tenha que adivinhar uma palavra que será mostrada com as letras embaralhadas. O programa terá uma lista de palavras lidas de um arquivo texto e escolherá uma aleatoriamente. O jogador terá seis tentativas para adivinhar a palavra.
