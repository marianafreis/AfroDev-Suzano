Tratando a coluna salário dos sonhos

```python
def tratamento_salario(elemento):
  resultado = elemento.replace('k+', '000').replace('k', '000').replace('mil', '000').replace(' ', '').replace('.', '').replace('mensais','').replace('reais','').replace('R$\xa0','').replace(',00', '')

  return int(resultado)

salario_sonhos = np.array([tratamento_salario(i) for i in salario_sonhos])
```

```python
salario_sonhos
```

Sabendo que o salário mínimo é de R$ 1.192,40, e o crescimento anual é de 5%, quantos anos cada pessoa levaria para chegar no salário dos sonhos, levando em consideração a coluna Qual seu salário dos sonhos 

```python
def anos_salario_sonhos(salario):
  salario_minimo = 1192.40
  anos = 0

  while salario_minimo < salario:
    anos += 1
    salario_minimo += salario_minimo * 0.05
  return anos
```

```python
array_sonhos = np.array([anos_salario_sonhos(salario) for salario in salario_sonhos])
print(array_sonhos)
```

```python
#total de anos de idade para alcançar o salário dos sonhos
total_anos = np.array(array_sonhos) + np.array(idades)
total_anos
```

```python
#id e total de anos
id_anos = np.array([[x+1, y] for x, y in enumerate(total_anos)])
id_anos
```

Separando quem não vai chegar ao salário dos sonhos antes dos 90 anos

```python
[[x, y] for x, y in id_anos if y >= 90]
```

Montando grupos de alunos por idade

```python
def grupos_idade():
  grupo_idade = {'menor_20': 0, 'maior_20_menor_30': 0, 'maior_30': 0}
  for idade in idades:
    if idade <= 20:
      grupo_idade.update({'menor_20': grupo_idade['menor_20'] + 1})
    elif idade > 20 and idade <= 30:
      grupo_idade.update({'maior_20_menor_30': grupo_idade['maior_20_menor_30'] + 1})
    else:
      grupo_idade.update({'maior_30': grupo_idade['maior_30'] + 1})

  return grupo_idade

grupo_alunos = grupos_idade()
```
