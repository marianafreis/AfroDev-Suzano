Lista de dicionários com o id e o nome do usuário:

```python
 users = [
  { "id": 0, "name": "Hero" },
  { "id": 1, "name": "Dunn" },
  { "id": 2, "name": "Sue" },
  { "id": 3, "name": "Chi" },
  { "id": 4, "name": "Thor" },
  { "id": 5, "name": "Clive" },
  { "id": 6, "name": "Hicks" },
  { "id": 7, "name": "Devin" },
  { "id": 8, "name": "Kate" },
  { "id": 9, "name": "Klein" }
]
```

Lista das amizades

```python
friendships = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4), 
  (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

Criando a lista de amigos de cada usuário

```python
for user in users:
  user["friends"] = []
```

```python
users
```

Adicionando os amigos à lista de amigos de cada usuário

```python
for i, j in friendships:
  if i == int(users[i]['id']):
    users[i]['friends'].append(users[j])
    users[j]['friends'].append(users[i])
```

```python
users
```

Criando um contador para calcular a média de amigos dos usuários

```python
count = 0

for x in users:
  count = count + len(x["friends"])
  #print(count)

media = count / len(users)
print(media)
```

Mostrando o número de amigos de cada usuário

```python
for user in users:
  print('{} tem {} amigos'.format(user['name'], len(user['friends'])))
```

Mostrando quais são os usuários que tem mais amigos

```python
max_friends = max([len(user['friends']) for user in users])
for user in users:
  if max_friends == len(user['friends']):
    print(user['name'])
```

Mostrando quais usuários tem menos amigos

```python
min_friends = min([len(user['friends']) for user in users])
for user in users:
  if min_friends == len(user['friends']):
    print(user['name'])
```

Mostrando quem são os amigos dos usuários mais populares

```python
max_friends = max([len(user['friends']) for user in users])
for user in users:
  if max_friends == len(user['friends']):
    print('{} -> {}'.format(user['name'], [friend['name'] for friend in user['friends']]))
```

Criando uma lista de tuplas com o id do usuário e a quantidade de amigos que ele tem

```python
lista = []

for user in users:
  lista.append((user['id'], len(user['friends'])))

lista
```
